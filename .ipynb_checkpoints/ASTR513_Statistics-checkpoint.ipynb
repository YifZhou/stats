## Lecture 2: Random Number Generators (Jan 13th)

### Characteristics of Random Number
* Criterion I: look at the distribution of the random number, it should be a unifrom distribution.
* Criterion II: we don't want any correlation between the sequence of the random number.
    >if we pick 4 random number once, and form a point in 4D space, we don't want the point being clustered; if we pick N random number once and sum it, the distribution of the sums would be Gaussian (larger N steeper peak)
* Dr. Knuth proposaled a way to test if a number sequence is random (enough?)
* Berford's Law: if a probability distribution P(x) is scale free (has nothing to do with units), then it can only be the random number, P(x) = const/x
    >see http://u.arizona.edu/~dpsaltis/Astr513/

### Linear congruential generator
$X_{n+1} = (a X_n + C)$ mod $m$

a is multiplier, $0<a<m$; c is increment, $0\le c<m$; m is modulus, $m>0$

* There is a period and such $P<m$. The largest m we can use is $2^{32}$. It can be proved that the number of random number is $N^2 \ll P = m$.
* Clustering effect: $k \rightarrow m^{1/k}$, if $k=5$, $2^{32/5}\sim 80$ (k: len(array) you pick from sequence)
* Efficiency
* Repeatability. You should be able to produce the exact same sequence of random numbers
* Portability. The random number sequence doesn't depend on devices.

Typicall you will want m to be the largest prime you can store, and a be a relatively large prime number to obtain the large period.

Beyond LGM:
* Non-linear methods. $X_n = (a X_{n-1}^2 + b X_{n-1} + c)$ mod $m$
* Multiple-recurssive methods (MRM). $X_n = (a_1 X_{n-1} + \cdots + a_k X_{n-k})$ mod $m$

### What if we want a likelihood distribution P(y)?
$$P(x)dx \rightarrow P(y)dy$$
let $P(x) = 1$ and $x \in [0, 1]$, we have
$$|P(x)dx| = |P(y)dy| \Rightarrow |dx| = P(y)|dy|$$
Just remove the absolute value symbol
$$dx = P(y)dy \Rightarrow x = \int P(y)dy$$
#### Exponential
$$P(y) = \frac{1}{a}e^{-y/a}\ y>0$$
$$x = \int \frac{1}{a}e^{-y/a} dy \Rightarrow x = -e^{-y/a}$$
$$\therefore y = -a \ln{x}$$

```python
plt.rcParams.update(newparams_for_screen)
a = 1
x = np.random.random(1000000)
y = - a*np.log(x)

# the histogram of the data
n, bins, patches = plt.hist(y, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
#y = mlab.normpdf( bins, mu, sigma)
#l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('y')
plt.ylabel('Likelihood')
plt.title(r'$\mathrm{Histogram\ of\ y:}\ a=1$')
plt.yscale('log', nonposy='clip')

plt.grid(True)

plt.show()
```

#### Gaussian Distribution
$$|P(y_1,y_2)dy_1dy_2| = |P(x_1,x_2)dx_1dx_2| \Longrightarrow P(y_1,y_2)=|\partial(x_1,x_2)/\partial(y_1,y_2)|$$

one way to construct the Gaussian is (refer to https://en.wikipedia.org/wiki/Box–Muller_transform)

Suppose U1 and U2 are independent random variables that are uniformly distributed in the interval (0, 1). Let
$$Z_0 = R \cos(\Theta) =\sqrt{-2 \ln U_1} \cos(2 \pi U_2)\,$$
and
$$Z_1 = R \sin(\Theta) = \sqrt{-2 \ln U_1} \sin(2 \pi U_2).\,$$
Then Z0 and Z1 are independent random variables with a standard normal distribution.

The derivation[4] is based on a property of a two-dimensional Cartesian system, where X and Y coordinates are described by two independent and normally distributed random variables, the random variables for R2 and Θ (shown above) in the corresponding polar coordinates are also independent and can be expressed as
$$R^2 = -2\cdot\ln U_1\,$$
and
$$\Theta = 2\pi U_2. \,$$
Because R2 is the square of the norm of the standard bivariate normal variable (X, Y), it has the chi-squared distribution with two degrees of freedom. In the special case of two degrees of freedom, the chi-squared distribution coincides with the exponential distribution, and the equation for R2 above is a simple way of generating the required exponential variate.

```python
def GenerateGaussian(n):
    x1 = np.random.random(n)
    x2 = np.random.random(n)
    two_pi = 2*np.pi
    y1 = np.sqrt(-2*np.log(x1))*np.cos(two_pi*x2)
    y2 = np.sqrt(-2*np.log(x1))*np.sin(two_pi*x2)
    return y1

plt.rcParams.update(newparams_for_screen)
fig = plt.figure()
ax = fig.add_subplot(111)

y = GenerateGaussian(1000000)
# the histogram of the data
n, bins, patches = plt.hist(y, 50, normed=0, facecolor='green', alpha=0.75)
ax.set_xlabel('y')
ax.set_ylabel('Counts')
ax.set_title(r'$\mathrm{Histogram\ of\ y}$')
#ax.yscale('log', nonposy='clip')

plt.show()
```

##### Berry–Esseen theorem
**Some people will use this to generate the Gaussian distribution, but it is dangerous than the previous one**

In probability theory, the central limit theorem states that, under certain circumstances, the probability distribution of the scaled mean of a random sample converges to a normal distribution as the sample size increases to infinity. Under stronger assumptions, the Berry–Esseen theorem, or Berry–Esseen inequality, gives a more quantitative result, because it also specifies the rate at which this convergence takes place by giving a bound on the maximal error of approximation between the normal distribution and the true distribution of the scaled sample mean.
One version, sacrificing generality somewhat for the sake of clarity, is the following:

There exists a positive constant C such that if X_1, X_2, ..., are i.i.d. random variables with $E(X_1) = 0$, $E(X_1^2) = σ^2 > 0$, and $E(|X_1|^3) = ρ < ∞$, and if we define
$$Y_n = {X_1 + X_2 + \cdots + X_n \over n}$$
the sample mean, with Fn the cumulative distribution function of
$${Y_n \sqrt{n} \over {\sigma}},$$
and Φ the cumulative distribution function of the standard normal distribution, then for all x and n,
$$\left|F_n(x) - \Phi(x)\right| \le {C \rho \over \sigma^3\,\sqrt{n}}.$$

Illustration of the difference in cumulative distribution functions alluded to in the theorem.
That is: given a sequence of independent and identically-distributed random variables, each having mean zero and positive variance, if additionally the third absolute moment is finite, then the cumulative distribution functions of the standardized sample mean and the standard normal distribution differ (vertically, on a graph) by no more than the specified amount. Note that the approximation error for all n (and hence the limiting rate of convergence for indefinite n sufficiently large) is bounded by the order of n−1/2.

### Maxwell-Boltzmann Distribution
The Maxwell–Boltzmann distribution is the function
$$f(v) = \sqrt{\left(\frac{m}{2 \pi kT}\right)^3}\, 4\pi v^2 e^{- \frac{mv^2}{2kT}}, $$
let $v' = \sqrt{m/kT} v$, we got
$$f(v) = \frac{2}{\sqrt{\pi}} v'^2 e^{-v'^2/2}$$
Also, for each component, 
$$f(v_x) = \frac{1}{\sqrt{2\pi}} e^{-v_x^2/2}$$

```python
n = 1000000
v = np.sqrt(np.sum([(GenerateGaussian(n))**2 for x in range(3)], axis=0))

plt.rcParams.update(newparams_for_screen)
fig = plt.figure()
ax = fig.add_subplot(111)

# the histogram of the data
n, bins, patches = plt.hist(v, 100, normed=0, facecolor='green', alpha=0.75)
ax.set_xlabel('v')
ax.set_ylabel('Counts')
ax.set_title(r'$\mathrm{Histogram\ of\ v}$')
#ax.yscale('log', nonposy='clip')

plt.show()
```

## Lecture 6 (Feb 24)

Assume you throw a object which only gravity applis on it. The object's trajectory is parabolic with height h and distance R. The initial velocity is u(ux, uy), g = 10m/s^2.
$$h = \frac{1}{2g} u_y^2$$
$$R = \frac{2u_xu_y}{g}$$

So 
$$u_x = \sqrt{\frac{g}{8}}\frac{R}{\sqrt{h}}$$
$$u_y = \sqrt{2gh}$$

If $h = 1\pm 0.2$m, and $R = 10\pm 0.2$m. Let's use Monte-Caro to generate them and plot at ux-uy plane.

```python
def GenerateGaussian(n, mean, sigma):
    x1 = np.random.random(n)
    x2 = np.random.random(n)
    two_pi = 2*np.pi
    y1 = np.sqrt(-2*np.log(x1))*np.cos(two_pi*x2)
    y2 = np.sqrt(-2*np.log(x1))*np.sin(two_pi*x2)
    return y1*sigma+mean

h = GenerateGaussian(1e6, 1, 0.2)
R = GenerateGaussian(1e6, 10, 0.2)
ux = np.sqrt(10/8)*R/np.sqrt(h)
uy = np.sqrt(2*10*h)
```

```python
plt.rcParams.update(newparams_for_screen)
fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_title(r"$h$,$R$ has Gaussian error, but $u_x$,$u_y$ are strong corrlated")
ax.set_xlabel(r"$u_x$")
ax.set_ylabel(r"$u_y$")

ax.scatter(ux, uy)
ax.plot(np.sort(ux), np.sort(4*h/R*ux)[::-1], color="green", lw = 3)
plt.show()
```

There is another way of doing it. 
$$ P(R, h) dR dh = P(u_x, u_y) du_x du_y $$
where $P(R,h) = G(R; R_0, \sigma_R)G(h; h_0, \sigma_h)$ is Gaussian distribution. In such case
$$ P(u_x, u_y) = G(R; R_0, \sigma_R)G(h; h_0, \sigma_h) J(\frac{R,h}{u_x,u_y})$$

$J(...)$ is the Jacobian Determinant
$$J(\frac{R,h}{u_x,u_y}) = \frac{\partial R}{\partial u_x}\frac{\partial h}{\partial u_y} - \frac{\partial h}{\partial u_x}\frac{\partial R}{\partial u_y} = \frac{\partial R}{\partial u_x}\frac{\partial h}{\partial u_y} $$
where $\frac{\partial h}{\partial u_x} = 0$

So we have
$$ P(u_x, u_y) = \frac{u_y^2}{\pi \sigma_R \sigma_h g^2} Exp\left[-\frac{(R-R_0)^2}{2\sigma_R^2}\right] Exp\left[-\frac{(h-h_0)^2}{2\sigma_h^2}\right] $$

The variable that we don't care is called nuisance parameter. Here when we want $P(u_y)$, $u_x$ is nuisance parameter.
$$ P(u_y) = \int P(u_x, u_y) du_x = \frac{u_y}{\sqrt{2\pi} \sigma_h g} Exp\left[-\frac{\left(\frac{u_y^2}{2g} - h_0\right)^2}{2\sigma_h^2} \right] $$

$$P(A\& B) = P(A)P(B|A)$$
$$P(B\& A) = P(B)P(A|B)$$
So
$$P(A|B) = \frac{P(A)P(B|A)}{P(B)}$$
However, if finding out $P(A)P(B|A)$ is eaiser, then you can just renormalize it to obtain $P(A|B)$.

So $P$(reality|measurement) = $C$ $P_r$(reality) $P$(measurement|reality). Put our example into this,
$$ P(u_x, u_y | R, h) = C P_r(u_x)P_r(u_y) P(R,h) = C P_r(u_x)P_r(u_y) G(R; R_0, \sigma_R)G(h; h_0, \sigma_h) $$
Just assume the prior $C P_r(u_x)P_r(u_y)$ is constant. Therefore
$$ P(u_y) = C' \frac{1}{u_y} Exp\left[-\frac{\left(\frac{u_y^2}{2g} - h_0\right)^2}{2\sigma_h^2} \right]$$
It weights more on lower end because of $\frac{1}{u_y}$ (see blue line above)

## Lecture 7 (Feb 29)

Pearson's $\chi^2$ test

N data point with ssumptions:
* uncertainties have $\sigma_x << \sigma_y$
* all uncentainties in y are Gaussian
* all data points are uncorrelated

Now we have a model $y = y(x; \vec{m})$. Since the error is Gaussian, so the value our model have at $x_i$ has a likehood of
$$ P_i = \frac{1}{\sqrt{2\pi \sigma_i^2}} Exp\left[ -\frac{\left(y(x;\vec{m})-y_i\right)^2}{2\sigma_i^2} \right]$$

So the final likehood is
$$ P = \prod_{i=1}^N \left( \frac{1}{\sqrt{2\pi \sigma_i^2}} Exp\left[ -\frac{\left(y(x;\vec{m})-y_i\right)^2}{2\sigma_i^2} \right]  \right)$$
$$ = \left( \prod_{i=1}^N \frac{1}{\sqrt{2\pi \sigma_i^2}}  \right) Exp\left[ \sum_{i=1}^N -\frac{\left(y(x;\vec{m})-y_i\right)^2}{2\sigma_i^2} \right] $$
Now we just need to find out the maximum of this value to fit the model to the data. Fitting SED sometimes uses this method.

More practically, assume $y=a+bx$, then 
$$\chi^2 = \sum_{i=1}^N \frac{\left[y_i - (a+bx)\right]^2}{\sigma_i^2} $$. To obtain the a and b that maximum value requires.

$$\frac{\partial \chi^2}{\partial a} = 0 \Rightarrow \sum\frac{y_i}{\sigma_i^2}-a\sum\frac{1}{\sigma_i^2}-b\sum\frac{x_i}{\sigma_i^2} = 0$$
$$\frac{\partial \chi^2}{\partial b} = 0 \Rightarrow \sum\frac{y_i x_i}{\sigma_i^2} - a\sum\frac{x_i}{\sigma_i^2}-b\sum\frac{x_i^2}{\sigma_i^2} = 0$$

simplify the notations, we can write down
$$S_y - a S - b S_x = 0$$
$$S_{xy} - a S_x - b S_{xx} = 0$$

$$ \Longrightarrow a = \frac{S_{xx}S_y - S_xS_{xy}}{\Delta} \ \ \ b = \frac{S S_{xy} - S_x S_y}{\Delta}$$
where $\Delta = S\cdot S_x - (S_x)^2$.

Nowadays, people use the following formula
$$ t_i = \frac{1}{\sigma_i}\left[x_i-\frac{S_x}{S}\right]$$
$$ S_{tt} = \sum_{i=1}^{N} t_i^2 $$
$$ b = \frac{1}{S_{tt}} \sum_{i=1}^{N} \frac{t_i y_i}{\sigma_i} $$
$$ a = \frac{S_y - S_x b}{S} $$

```python
def getab():
    # Generate linear data with errors
    x = 3 * np.random.random_sample((10)) + 1
    x = np.sort(x)
    y = 1 + 2*x
    sigma = 0.1
    error = GenerateGaussian(10, 0, sigma)
    y = y + error

    Sx = np.sum(x/sigma**2)
    Sy = np.sum(y/sigma**2)
    S = len(x)/sigma**2

    t = 1/sigma * (x - Sx/S)
    Stt = np.sum(t**2)

    b = 1/Stt * np.sum(t*y/sigma)
    a = (Sy - Sx*b)/S
    return a, b

n = 1e5
a = np.zeros(n); b = np.zeros(n)
for i in range(int(n)):
    a[i], b[i] = getab()
    

plt.rcParams.update(newparams_for_screen)
n, bins, patches = plt.hist(a, 50, normed=1, facecolor='green', alpha=0.75)
plt.xlabel("a")
plt.show()

plt.rcParams.update(newparams_for_screen)
n, bins, patches = plt.hist(b, 50, normed=1, facecolor='green', alpha=0.75)
plt.xlabel("b")
plt.show()

fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.scatter(a, b)
ax.set_xlabel("a"); ax.set_ylabel("b")
plt.show()
```

Such anti-correlation is due to 
$$ a = \frac{S_y}{S} - \frac{S_x}{S} b$$
